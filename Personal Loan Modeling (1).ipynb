# **📊 Outlier Detection (IQR Method) — Without Row Removal**

In this step, I analyzed potential outliers in key numeric columns such as **Income**, **Mortgage**, and **CCAvg** using the **Interquartile Range (IQR)** method.  
The goal was to identify **unusually high or low values** that could influence the analysis while ensuring the dataset’s **real-world accuracy** is maintained.


```python
iqr_lst = ['Income','Mortgage','CCAvg']
```

```python
df.describe().T
```

```python
for i in iqr_lst:
    Q1 = df['Income'].quantile(0.25)
    Q3 = df['Income'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 *IQR
    upper_bound = Q3 +1.5 * IQR
    print(df[(df['Income'] < lower_bound) | (df['Income'] > upper_bound)][i])
```

## 🔍 Observations

- Columns like **Income**, **CCAvg**, and **Mortgage** showed some extreme values, but they appear **valid and realistic**.  
- For example, the **Income** column contained a few customers with **very high annual incomes**, which aligns with real-world scenarios in the banking sector.  
- According to the **Kaggle dataset documentation**, this dataset is based on **authentic banking data**, and such variations represent **natural customer diversity**.  
- Hence, **no rows were removed or replaced**, as doing so could **distort the actual business insights** and lead to **biased analysis**.

---

## 💡 Conclusion

Although the **IQR method** flagged certain data points as potential outliers, these values are **genuine and significant** for understanding real-world customer patterns.  
Therefore, I chose to **retain all records** to preserve the dataset’s integrity.
